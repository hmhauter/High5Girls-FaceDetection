# Testing of the Recognition

This script tests if our trained NN works
1. It uses the saved NN (trainer/trainer.yml)
2. Add your names accirding to the IDs to the name list (first name is ID=0, second name is ID=1, ...)
3. Start the script and see if it can detect AND NOW ALSO RECOGNIZE your faces
4. Test together with friends 


```python
###### ADD NAMES HERE ############
# ID=0, ID=1, ID=2, ...
names = ['Name1', 'Name2'] 
```

```python
# again imports
import cv2
import numpy as np
```

```python
# get our NNs
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read('./trainer/trainer.yml')
cascadePath = "../haarcascade_frontalface_default.xml"
faceCascade = cv2.CascadeClassifier(cascadePath)

# some pretty stuff..
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
# again we are starting the laptop camera
cam = cv2.VideoCapture(0)
cam.set(3, 640) # set video widht
cam.set(4, 480) # set video height

# define min window size to be recognized as a face
minW = 0.1*cam.get(3)
minH = 0.1*cam.get(4)

while True:
    # get image from laptop camera
    ret, img =cam.read()
    # convert tto gray scale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # DETECT faces in image
    faces = faceCascade.detectMultiScale( 
        gray,
        scaleFactor = 1.2,
        minNeighbors = 5,
        minSize = (int(minW), int(minH)),
       )

    for(x,y,w,h) in faces:

        cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 2)
        # RECOGNIZE the face from the detection
        id, confidence = recognizer.predict(gray[y:y+h,x:x+w])
        print("DETECTED ID: ", id)
        print("WITH CONFIDECE: ", confidence)

        # Check if confidence is less them 100 ==> "0" is perfect match 
        if (confidence < 100):
            id = names[id]
            confidence = "  {0}%".format(round(100 - confidence))
        else:
            id = "unknown"
            confidence = "  {0}%".format(round(100 - confidence))
        
        cv2.putText(img, str(id), (x+5,y-5), font, 1, (255,255,255), 2)
        cv2.putText(img, str(confidence), (x+5,y+h-5), font, 1, (255,255,0), 1)  
    
    cv2.imshow('camera',img) 

    k = cv2.waitKey(10) & 0xff # Press 'ESC' for exiting video
    if k == 27:
        break

# Do a bit of cleanup
print("\n [INFO] Exiting Program and cleanup stuff")
cam.release()
cv2.destroyAllWindows()
```
